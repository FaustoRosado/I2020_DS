FFANG Stock Market Analysis During NYC Lockdown 

Events in the past have affected the price of stocks after occurring, i.e. subprime crash of 2008, Japanese tsunami in 2011 and more recently Covid related events and GME short sale. 
For this research we chose to pull stock data from Yahoo Finance to find how stocks were affected by Covid lockdowns and relief/reopening periods. We chose to use the major tech stocks, FAANG stocks- Facebook, Amazon, Apple, Netflix and Google, together comprise 15% of the S&P 500 - like 15% of the US economy. 
For the Covid events, we chose to use NYC lockdown date and announcement for restaurant capacity to increase to 25% as an indication of Covid reopening event. 

We chose to observe analyze 30 days after each Covid event and the impact on the stock prices.
From the data observed, we found a period of 30 days was sufficient to capture stock swings. 
Our hypothesis is that we expect a similar increase across FAANG stocks after Covid lockdown and an a similar decrease over the same period after the early indication of Covid reopening as defined above.

Though Covid is a special unique event, we expect to show linear relationship before and after lockdown. 
Will there be similar swings noted on all 5 stocks after each period? 
For the period after lockdown when people remained home FAANG stocks were a good cross section/representation of social media, e-commerce, tech hardware, entertainment & tech conglomerate. These companies dominate these areas.
Reason for the Report:
The stock market provides a view of what investors expect for the future. We can look to the stock market for complex events like Covid-19 to gauge what investors think, but also get a view of future behavior. 
It’s a bet on the future or speculating on the direction the market on that day. 
Because of the all uncertainty during Covid, we decided to use NYC lockdown date, We chose 30 day trading time frame, good period we think lockdown will sink in. 
We expect a similar increase across all five stocks and people will adjust to staying and working from home.
Likewise, we chose February 12, 2021 as end of lockdown, 25% indoor dining in restaurants and venues with over 10,000 seats could reopen at 10% capacity. Again 30 day trading time frame, good sink in period to gauge stock swing.

Conclusion:
Our findings supported our hypothesis during the period analysis after lockdown as expected. Speculation that these stocks would rise were in line with people’s ideas that working and living from home would increase people’s reliance on these tech companies for social media, commerce, hardware purchases, entertainment and all of Google services.
On the other hand, picking a date for reopening is tough since we are still in this period. Though there was not a similar decrease in stock prices, 3 out of 5 did experience a down swing, in line with people shopping more outside watching less TV, in the case of Netflix. The surprise here is Apple, since it operates real stores. 
The slight upswing with Facebook & Google could be explained by their dominance in social media and google services on all devices, especially mobile. 
We think as the economy and society resemble normal life after Covid, we may see similar downward swings in the stock price of these five companies.
